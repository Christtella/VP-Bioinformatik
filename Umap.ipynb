```
#r "nuget: FSharp.Stats, 0.4.12-preview.2"
#r "nuget: Plotly.NET"
#r "nuget: Plotly.NET.Interactive"
#r "nuget: Deedle, 3.0.0"
#r "nuget: Deedle.Interactive, 3.0.0"
#r "nuget: FSharp.Data"

open FSharp.Data
open FSharp.Stats
open System
open Plotly.NET
open Plotly.NET.Interactive
open Deedle
open Plotly.NET.TraceObjects
```

```
let f = Frame.ReadCsv("/Users/christelsakhel/Desktop/VP1 bioinformatik/TrainNewFrame.csv")
f
```

```
let numberHeader = 
    f.ColumnKeys 
    |> Array.ofSeq
    |> Array.filter (fun x -> x.Contains "numbers")
```

```
type ProteinInfo = {
    Key:int
    Input: string
    Location: string
    Membrane: string
    Embedding: float32[]
    }    
```

```
let result =
    f
    |> Frame.mapRows (fun k werte -> 
            {
                Key = werte.GetAs<int>("Key")
                Input = werte.GetAs<string>("input")
                Location = werte.GetAs<string>("loc")
                Membrane = werte.GetAs<string>("membrane")
                Embedding = 
                    numberHeader
                    |> Array.map (fun header -> 
                        werte.GetAs<float32>(header)
                        )
            }
        )
    |> Series.values
    |> Array.ofSeq
```

```
let justEmbedding : float32[][] =
    result
    |> Array.map (fun x -> x.Embedding)
justEmbedding
```

```
#r "nuget: UMAP, 1.0.34015"
open UMAP
```

```
// It doesn't matter where this data comes from, so long as it is a
// float[][] and every nested array has the same length
//let vectors : float32[][] = ..

// Calculate embedding vectors using the default configuration

let umap = Umap()
let numberOfEpochs = umap.InitializeFit(justEmbedding)
for i in 0 .. numberOfEpochs - 1 do
    umap.Step() |> ignore

// This will be a float[][] where each nested array has two elements
// because the default Umap configuration generates 2D embeddings
let embeddings = umap.GetEmbedding()
embeddings
```

```
let xData = 
    embeddings
    //|> Array.toList
    |> Array.map (fun x -> x.[0])
```

```
let yData = 
    embeddings
    //|> Array.toList
    |> Array.map (fun x -> x.[1])
```

```
let justLocation =
    result
    |> Array.map (fun x -> x.Location)
justLocation
```

```
let colorMapping = dict [("Cell.membrane", Red); ("Cytoplasm", Blue); ("Endoplasmic.reticulum", Green); 
    ("Golgi.apparatus", Yellow); ("Lysosome/Vacuole", Orange); ("Mitochondrion", Purple); 
    ("Nucleus", Pink);("Peroxisome", Cyan); ("Plastid", Magenta); ("Extracellular", Brown)]
```

```
let colors = 
    justLocation
    |> Array.map (fun loc -> colorMapping.[loc]) 
colors
```

```
let myEmbeddedChart = 
    Array.mapi2 (fun i x y ->
        Chart.Point([x,y], MarkerColor = Color.fromKeyword colors.[i])) xData yData
    |> Chart.combine
    //Chart.Point(xData, yData, )
    //|> MarkerColor = Color.fromColors colors
    |> Chart.withLegend(false)
    |> Chart.withTitle "Embeddings into UMAP"
    |> Chart.withXAxisStyle ("UMAP1")
    |> Chart.withYAxisStyle ("UMAP2")
    
    
myEmbeddedChart
```

**Alternative**

```
type LocationType  = {
    Location: string array
    X : float32 array
    Y : float32 array
    Color: string
    }    

let NachColors =
            {
                Location = justLocation
                X = xData
                Y = yData

            }
NachColors
```

```
let mapLocationToColor location =
    match location with
    | "Cell.membrane" -> "Red"
    | "Cytoplasm" -> "Blue"
    | "Endoplasmic.reticulum" -> "Green"
    | "Golgi.apparatus" -> "Yellow"
    | "Lysosome/Vacuole" -> "Orange"
    | "Mitochondrion" -> "Purple"
    | "Nucleus" -> "Cyan"
    | "Peroxisome" -> "Magenta"
    | "Plastid" -> "Brown"
    | "Extracellular" -> "Pink"
    | _ -> "Black"
```

```
let groupedLocations =
    Array.zip3 NachColors.Location NachColors.X NachColors.Y
    |> Array.groupBy (fun (location, _, _) -> location)
    |> Array.map (fun (location, values) ->
        {
            Location = [| location |]
            X = values |> Array.map (fun (_, x, _) -> x)
            Y = values |> Array.map (fun (_, _, y) -> y)
            Color = mapLocationToColor location
        }
    )
groupedLocations
```
